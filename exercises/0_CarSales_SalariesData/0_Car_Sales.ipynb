```python

#   c) Draw a line chart of average CO2 emissions for same years that as in b)
df_emissions_year = df.groupby("Year")["Avg_CO2"].sum()
df_emissions_year.plot(x="Year", y="Avg_CO2")
```

```python

#   d) Draw a line chart of all years and months for import
df_year_month_import = df.groupby(["Year", "Month"])["Import"].sum()
df_year_month_import.plot(y="Import", x=["Year","Month"])
```

```python

#   e) Draw a line chart of all years and months for average CO2 emissions
df_emissions_year_month = df.groupby(["Year", "Month"])["Avg_CO2"].sum()
df_emissions_year_month.plot(x=["Year", "Month"], y="Avg_CO2")
```

```python

#   f) Draw a line chart of all years and months for electric cars import where it's relevant.
df_electric_year_month = df[df["Import_Electric"] > 0]
df_electric_year_month = df_electric_year_month.groupby(["Year", "Month"])["Import_Electric"].sum() #.reset_index()
df_electric_year_month
# df_electric_year_month_relevant = df_electric_year_month.query("Import_Electric > 0")

df_electric_year_month.plot(x=["Year","Month"], y="Import_Electric")
```

```python

#   g) Draw a line chart of average diesel share per year
df_dieselshare_year = df.groupby("Year")["Diesel_Share"].mean()
df_dieselshare_year.plot(x="Year", y="Diesel_Share")
```

```python
df
# Quantity, Quantity_Diesel, Quantity_Hybrid, Quantity_Electric
df.columns
df_year_all_quantitys = df.groupby("Year")[["Quantity", "Quantity_Diesel", "Quantity_Hybrid", "Quantity_Electric"]].sum()
df_year_all_quantitys["Quantity_Bensin"] = (df_year_all_quantitys["Quantity"].fillna(0) - df_year_all_quantitys["Quantity_Diesel"].fillna(0) - df_year_all_quantitys["Quantity_Hybrid"].fillna(0) - df_year_all_quantitys["Quantity_Electric"].fillna(0))
df_year_all_quantitys_with_bensin = df_year_all_quantitys.reset_index()
df_year_all_quantitys_with_bensin
```

```python
df_year_all_quantitys_with_bensin = df_year_all_quantitys_with_bensin[df_year_all_quantitys_with_bensin["Year"] < 2017]
df_year_all_quantitys_with_bensin.plot(x="Year", y=["Quantity", "Quantity_Diesel", "Quantity_Hybrid", "Quantity_Electric", "Quantity_Bensin"])
```
