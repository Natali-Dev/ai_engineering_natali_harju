```python
# Profit per age group
# ordningen på raderna ändras varje gång man körs pga - den plockar fram det som är mest effektivt (snabbaste sättet att få ut datan)
# Groupa på det du INTE aggregerar! 
# Vid aggregering måste du ta HAVING, where fungerar inte
import duckdb
df_small_groups = duckdb.query(
    """
    SELECT 
    age_group, 
    country,
    customer_gender,
    sum(profit) as Total_Profit,
    count(*) as Number_Rows
    FROM df
    group by age_group, Country, customer_gender
    having Number_Rows < 1000
    order by Total_Profit desc
    
    """
).df()
df_small_groups
```

```python
df_senior_female = df_small_groups.query("Age_Group == 'Seniors (64+)' and Customer_Gender == 'F'")
df_senior_female
```

```python
# import matplotlib.pyplot as plt

# ax, bar = plt.Subplot(1)
#Vi kan plotta såhär för att pandas har byggt denna rapper av matplotlib! utan kind=bar blir lineplot
#barh = flippad barplot
ax = df_senior_female.plot(kind="barh", x="Country", y="Total_Profit")
ax.set(title="females 64+ total profit in bike store", ylabel="Country", xlabel="Total Profit")
ax.invert_yaxis() #BLIR snyggare
fig = ax.get_figure()
fig.tight_layout()
fig.savefig("Profit per country in females 64+.png", dpi=200)
```

## MED DUCKDB

```python
import duckdb 
df_profit = duckdb.query(
    """
    SELECT Country, mean(Profit) as Profit_mean, sum(Profit) as Profit_sum, median(Profit) as Profit_median
    FROM df
    GROUP BY Country
    ORDER BY Profit_sum desc
    """
).df()
df_profit
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(8,4))

ax.bar(x=df_profit["Country"], height=df_profit["Profit_sum"])
ax.set(xlabel= "Country", ylabel="Total profit", title= "Total profit per country")
```
